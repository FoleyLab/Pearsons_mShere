```python
import pandas as pd
import numpy as np
from scipy import stats


### get tab 2 - this is the tab that contains the query set
geo_data_cell_2 = pd.read_excel('GEO.xlsx', sheet_name=1)
### empty array to store the gene names from tab 2
### this is the data that defines the query set
query_names = []

### get tab 1 - this is the tab that contains the data associated
### with the query set + many more genes... trick is we need
### to pick-out only the elements of this data set that correspond
### to the query set
geo_data_cell_1 = pd.read_excel('GEO.xlsx', sheet_name=0) 

### empty array to store the indices of the genes in the
### data set that are also in the query set
query_indices = []

### We will deal with the query set first... in particular,
### let's figure out where the C1 genes end and the C2 genes begin.
### We will consider the C1 genes only at this stage

# loop through all elements of the query set
for i in range(0,len(geo_data_cell_2)):
    # get the gene name
    gene_label = geo_data_cell_2.iloc[i,0]
    # determine of the second character of the gene name is a 
    # 2; if it is, we will stop reading there!
    if int(gene_label[1])==2:
        divider_t2 = i-1
        break
    # store the C1 gene names to the query_names array
    query_names.append(str(gene_label))
    
print(query_names)
print(len(query_names))
### Now that we have the query set stored in query_names, 
### let's go through the data set and identify the locations of 
### genes from the query set

# index for the query_names list... start at the top!
qidx = 0
for i in range(0,len(geo_data_cell_1)):
    # get the gene name of the gene from the data set
    gene_label = geo_data_cell_1.iloc[i,0]
    # check to see if this gene matches the query_names
    if query_names[qidx] == str(gene_label):
        # if yes, append it to the query_indices list
        query_indices.append(i)
        # advance the index used by query_names so that we
        # go to the next gene from the query set
        qidx += 1
    # check to see if we have moved from the C1 to C2 genes
    if int(gene_label[1])==2:
        # note the index of the last C1 gene
        divider_t1 = i-1
        break

### Get some basic information from the above:
# how many query set matches did we find in the data set?
print(len(query_indices))

# can we confirm a few matches?
print("Now testing that we sorted correctly")
print("First five elements of the query set")
print(query_names[0:5])
print("Matching items from the data set")
print(geo_data_cell_1.iloc[query_indices[0:5], 0])




```
